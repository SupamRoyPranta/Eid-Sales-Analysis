# Exploratory Data Analysis

# Gender

```python
Gender = df['Gender'].value_counts()

plt.pie(
    Gender,                              
    labels=Gender.index,                 
    autopct='%1.1f%%',                   
    startangle=90,                       
)

# Step 3: Add title
plt.title('Distribution of Male and Female')
plt.show()
```

```python
# Analysing Gender by Total Sales

Sales=df.groupby(['Gender'],as_index=False)['Amount'].sum()
ax=sea.barplot(x='Gender',y='Amount',data=Sales,hue='Gender')

for bar in ax. containers:
    ax.bar_label(bar, labels=[f'{value.get_height()/1_000_000:.1f}M' for value in bar], fontsize=12)

plt.title('Total Sales by Gender', fontsize=16)
plt.ylabel('Amount (in millions)', fontsize=12)
```

# Marital Status

```python
# Analysing Marital Status by Total Sales

sea.set(rc={'figure.figsize':(8,4)})

Sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum()
ax=sea.barplot(x='Marital_Status',y='Amount',data=Sales,hue='Gender')

for bar in ax. containers:
    ax.bar_label(bar, labels=[f'{value.get_height()/1_000_000:.1f}M' for value in bar], fontsize=10)

plt.title('Total Sales by Marital Status', fontsize=16)
plt.ylabel('Amount (in millions)', fontsize=12)

```

# Age Group 

```python
ax=sea.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Age Group by Gender')    
```

```python
# Age group by Sales
Sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax=sea.barplot(x='Age Group',y='Amount', hue='Age Group',data=Sales)

plt.title('Age Group by Total Sales')
```

# Occupation 

```python

sea.set(rc={'figure.figsize':(20,6)})
ax=sea.countplot(x='Occupation',hue='Occupation',data=df)
for bar in ax. containers:
    ax.bar_label(bar)  
plt.title('Number of Customer by Occupation',fontsize= 25)    
```

```python
sea.set(rc={'figure.figsize':(20,6)})
Sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sea.barplot(x='Occupation',y='Amount',hue='Occupation',data=Sales)
plt.title('Total Sales by Occupation',fontsize= 25)    
```

# States

```python
Orders=df.groupby(['State'],as_index=False)['Orders'].sum()
ax=sea.barplot(x='State', y='Orders',data= Orders,hue='State')
for bar in ax. containers:
    ax.bar_label(bar)
plt.title('Numbers of Orders by States', fontsize=25)
```

```python
sea.set(rc={'figure.figsize':(18,6)})
Sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sea.barplot(x='State', y='Amount',data= Sales,hue='State')
for bar in ax. containers:
    ax.bar_label(bar, labels=[f'{value.get_height()/1_000_000:.1f}M' for value in bar], fontsize=12)

plt.title('Total Sales by States', fontsize=25)
plt.ylabel('Amount (in millions)', fontsize=12)
```

# Product Category 

```python
# Orders by Product Category
sea.set(rc={'figure.figsize':(20,5)})

Order=df.groupby('Product_Category',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax=sea.barplot(x='Product_Category',y='Orders',data=Order,hue='Product_Category')
for bar in ax. containers:
    ax.bar_label(bar)
plt.title('Numbers of Orders by Product Category', fontsize=25)
```
